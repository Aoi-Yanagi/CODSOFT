```python
# --- CELL 1: IMPORTS FOR ML ---
import pandas as pd
import re
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, classification_report
print("Imports loaded successfully!")
```

```python
# --- CELL 2: LOADING THE TEXT FILES INTO TABLES ---
folder_path = 'Dataset/'
study_material_path = folder_path + 'train_data.txt'
exam_answer_key_path = folder_path + 'test_data_solution.txt'

study_information_table = pd.read_csv(
    study_material_path, 
    sep=' ::: ', 
    engine='python', 
    names=['Movie_ID', 'Movie_Title', 'Actual_Genre', 'Story_Plot']
)

exam_information_table = pd.read_csv(
    exam_answer_key_path, 
    sep=' ::: ', 
    engine='python', 
    names=['Movie_ID', 'Movie_Title', 'Actual_Genre', 'Story_Plot']
)

print("Data loaded! Here is a peek at the first 3 rows of the table:")
display(study_information_table.head(3))
```

```python
# --- CELL 3: SEPARATING THE STORIES FROM THE GENRES ---
# What the model reads to learn (the plots)
learning_stories = study_information_table['Story_Plot']
testing_stories = exam_information_table['Story_Plot']

# What the model is trying to predict (the genres)
learning_goals = study_information_table['Actual_Genre']
exam_answer_key = exam_information_table['Actual_Genre']
print("Stories and genres successfully separated.")
```

```python
# --- CELL 3.5: THE DATA CLEANING ---
print("Preprocessing the stories to remove punctuation and numbers... (Please wait)")

def movie_plot(raw_story):
    clean_text = re.sub(r'[^a-zA-Z\s]', '', raw_story)
    return clean_text.lower()

# Running our stories through the preprocessing
learning_stories = learning_stories.apply(movie_plot)
testing_stories = testing_stories.apply(movie_plot)

print("Cleaning complete! The stories are clean.")
```

```python
# --- CELL 4: TRANSLATING WORDS TO NUMBERS ---
print("Translating words with an expanded dictionary... (Please wait)")

word_translator = TfidfVectorizer(
    stop_words='english', 
    max_features=40000,    # Expanded to hold more two-word phrases
    ngram_range=(1, 2),    # Reading single words and pairs
    sublinear_tf=True      # Tones down words that repeat too much
)

# Using the original, un-scrubbed stories so we keep numbers like "007" or "1984"
learning_stories_numbered = word_translator.fit_transform(learning_stories)
testing_stories_numbered = word_translator.transform(testing_stories)

print("Translation complete! The data is prepped.")
```

```python
# --- CELL 5: TRAINING THE MODEL ---
print("Training the Model... (This will take a moment)")

logic_model = LogisticRegression(C=3.0, max_iter=2500)
logic_model.fit(learning_stories_numbered, learning_goals)

print("Training complete! The Model is ready.")
```

```python
# --- CELL 6: TAKING THE EXAM & CHECKING THE SCORE ---
print("Taking the final test...")
predictions = logic_model.predict(testing_stories_numbered)

# Grading the test
final_score = accuracy_score(exam_answer_key,predictions)

print("-" * 50)
print(f"🏆 The Confident Logic Model scored an accuracy of: {final_score * 100:.2f}%")
print("-" * 50)
```
